## 3. Tokenizar

```python
# tokenizar
# Tokenizar. Obtener tokens y lemas. total = 350,737 tokens sin duplicados = 24,275 tokens
#tokens_dict["palabra"] = [FREC, POS, LEMMA]
tokens_dict = {}

for noticia in df['Contenido']:
  temp_noticia = nlp(noticia)
  aux = [token for token in temp_noticia]
  #df['tokens'] = aux
  for token in aux:
    if token.text in tokens_dict:
      tokens_dict[token.text] += 1
    else:
      tokens_dict[token.text] = 1
```

```python
# obtener diccionario de tokens
print(tokens_dict)
```

## 4. Implementar posting list con índice invertido

```python
# implementar posting list con índice invertido
#posting_list["palabra"] = [ID1, ID2, ...] donde IDn es el índice de la noticia n
posting_list = {}

for token in tokens_dict:
#  print(type(token))
#  print(token)
  for noticia in df['Contenido']: 
    #print(noticia)
    #print(type(noticia))
    if token in noticia:
      if token in posting_list: posting_list[token].append(df.loc[df['Contenido'] == noticia, 'index'].iloc[0])
      else: 
        posting_list[token] = []
        posting_list[token].append(df.loc[df['Contenido'] == noticia, 'index'].iloc[0])
```

```python
print(posting_list["que"])
print(len(posting_list["que"]))
```

El posting list con índice invertido nos dirá, por palabra o token, los IDs de las noticias que contienen dicho token en forma de lista...
